```python
#import libraries
import pandas as pd
import numpy as np
from dash import Dash, html, dcc
import plotly.express as px
import datetime as dt
```

```python
#uploading file
df = pd.read_csv('/Users/ghazian/Downloads/soal/dkatalis/retail-banking-demo-data/LuxuryLoanPortfolio.csv')
df.head()
```

```python
#create business metrics num 1, quarterly loan volume

df['quarter'] = pd.to_datetime(df['funded_date']).dt.quarter
df['year'] = pd.DatetimeIndex(df['funded_date']).year

avg_loan_value = df.groupby(['year','quarter']).agg({'loan_id':'nunique','funded_amount':'sum'}).reset_index().sort_values(['year','quarter'],ascending=True)
avg_loan_value.head()
```

```python
#create business metrics num 2, expected return
df['expected_return'] = df['funded_amount'] * df['interest rate'] * df['duration years']

expected_return = df.groupby(['year','quarter']).agg({'loan_id':'nunique','expected_return':'sum'}).reset_index().sort_values(['year','quarter'],ascending=True)
expected_return['expected_return'] = expected_return['expected_return'].astype('int')
expected_return.head()
```

```python
#create business metrics num 2, Cycle Stage Length in Month
cycle_stage = df.groupby(['year','quarter']).agg({'loan_id':'nunique','duration months':'sum'}).reset_index().sort_values(['year','quarter'],ascending=True)
cycle_stage['cycle_stage'] = (cycle_stage['duration months'] / cycle_stage['loan_id']).astype('float')
cycle_stage.head()
```
